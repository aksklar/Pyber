## Pyber

Your objective is to build a Bubble Plot that showcases the relationship between four key variables:

Average Fare ($) Per City, 
Total Number of Rides Per City, 
Total Number of Drivers Per City, 
City Type (Urban, Suburban, Rural)

In addition, you will be expected to produce the following three pie charts:

% of Total Fares by City Type, 
% of Total Rides by City Type, 
% of Total Drivers by City Type

```python
# Import Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read CSVs
city_csvpath = "raw_data/city_data.csv"
ride_csvpath = "raw_data/ride_data.csv"
city_data = pd.read_csv(city_csvpath)
ride_data = pd.read_csv(ride_csvpath)
```

```python
# Merge CSVs
pyber_data = pd.merge(city_data, ride_data, on=["city", "city"], how="right")
pyber_data = pyber_data[["city", "date", "fare", "ride_id", "driver_count", "type"]]
pyber_data.head()
```

```python
# Create bubble chart that showcase relationships between 
# Calculate the average fare ($) per city (y axis)
avg_fare = pyber_data.groupby(["city"]).mean()["fare"]
avg_fare.head()
```

```python
# Total number of rides per city (x axis)
total_rides = pyber_data.groupby(["city"]).count()["fare"]
total_rides.head()
```

```python
#city type & total drivers
cities_types = pyber_data.loc[:, ["city", "type", "driver_count"]]
cities_types = cities_types.drop_duplicates()
cities_df = cities_types.set_index("city")
cities_df["Average Fares"] = avg_fare
cities_df["Total Rides"] = total_rides
cities_df = cities_df.rename(columns={"type": "City Type", "driver_count": "Total Drivers"})
cities_df = cities_df[["Average Fares", "Total Rides", "Total Drivers", "City Type"]]
cities_df.head()
```

```python
with sns.axes_style("darkgrid"):
    sns.lmplot(data=cities_df, x="Total Rides", y="Average Fares", fit_reg=False, hue="City Type", palette=("lightcoral", "lightskyblue", "gold"), legend_out=False, size=6.5)
    
x_axis = cities_df["Total Rides"]
y_axis = cities_df["Average Fares"]
z_axis = cities_df["Total Drivers"]


plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare($)")
plt.xlim(-5, 70)
plt.ylim(15, 51)
plt.grid(True)


plt.scatter(x_axis, y_axis, s=z_axis*17, alpha=0.7, c=colors, edgecolor="black")
plt.show()
```

## % of Total Fares by City Type

```python
fare_city_type = pyber_data.groupby("type").sum()
fare_city_type
```

```python
total_fare = fare_city_type["fare"].sum()
rural_fare = fare_city_type.loc["Rural", "fare"]
rural_fare_percent = rural_fare / total_fare
```

```python
suburb_fare = fare_city_type.loc["Suburban", "fare"]
suburb_fare_percent = suburb_fare / total_fare
```

```python
urban_fare = fare_city_type.loc["Urban", "fare"]
urban_fare_percent = urban_fare / total_fare
```

```python
city_types = ["Rural", "Suburban", "Urban"]
fare_percent = [rural_fare_percent, suburb_fare_percent, urban_fare_percent]
colors = ["gold", "lightskyblue", "lightcoral"]
explode= (0, 0, 0.1)

plt.pie(fare_percent, explode=explode, labels=city_types, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=125)
plt.axis("equal")
plt.title("% of Total Fares by City Type")
fig = plt.figure(figsize=(15, 15))
plt.show()
```

## % Total Rides by City Type

```python
rides_city_type = pyber_data.groupby(["type"]).count()
rides_city_type
```

```python
total_rides = rides_city_type["ride_id"].sum()
total_rides
```

```python
rural_rides = rides_city_type.loc["Rural", "ride_id"]
rural_rides_percent = rural_rides / total_rides
rural_rides_percent
```

```python
suburb_rides = rides_city_type.loc["Suburban", "ride_id"]
suburb_rides_percent = suburb_rides / total_rides
suburb_rides_percent
```

```python
urban_rides = rides_city_type.loc["Urban", "ride_id"]
urban_rides_percent = urban_rides / total_rides
urban_rides_percent
```

```python
city_types = ["Rural", "Suburban", "Urban"]
rides_percent = [rural_rides_percent, suburb_rides_percent, urban_rides_percent]
colors = ["gold", "lightskyblue", "lightcoral"]
explode= (0, 0, 0.1)

plt.pie(rides_percent, explode=explode, labels=city_types, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=125)
plt.axis("equal")
plt.title("% of Total Rides by City Type")
fig = plt.figure(figsize=(15, 15))
plt.show()
```
